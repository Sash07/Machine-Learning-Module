```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset("iris")
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
sc.fit_transform(x_train)
```

```python
sc.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
svc=SVC()
```

```python
param_grid={"kernel":['linear', 'poly', 'rbf', 'sigmoid'],"gamma":["scale","auto"],"C":[10,20,30]}
```

```python
from sklearn.model_selection import GridSearchCV
grid_obj=GridSearchCV(svc,param_grid=param_grid,scoring="accuracy",cv=5)
```

```python
grid_obj.fit(x_train,y_train)
```

```python
grid_obj.best_params_
```

```python
grid_obj.best_estimator_
```

```python
y_pred=grid_obj.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

# SVR

```python
from sklearn.svm import SVR
df
```

```python
x=df.iloc[:,0:3]
x
```

```python
y=df.iloc[:,-2]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
sc.fit_transform(x_train)
sc.transform(x_test)
```

```python
svr=SVR()
```

```python
param_grid1={"kernel":['linear', 'poly', 'rbf', 'sigmoid'],"gamma":["scale","auto"],"epsilon":[0.1,0.3,0.5]}
```

```python
grid_obj1=GridSearchCV(svr,param_grid=param_grid1,cv=5,scoring="r2")
```

```python
grid_obj1.fit(x_train,y_train)
```

```python
grid_obj1.best_params_
```

```python
grid_obj1.best_estimator_
```

```python
y_pred=grid_obj1.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
score=mean_squared_error(y_pred,y_test) # should be minimum
```
