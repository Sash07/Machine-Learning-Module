```python
import numpy as np
def calculate_statistics(data):
    range_value = max(data) - min(data)
    variance_value = np.var(data, ddof=0)  # Population variance
    std_deviation = np.std(data, ddof=0)  # Population standard deviation
    
    return range_value, variance_value, std_deviation

# Example dataset
data = [10, 15, 20, 25, 30, 35, 40]

range_val, variance_val, std_dev_val = calculate_statistics(data)

print(f"Range: {range_val}")
print(f"Variance: {variance_val}")
print(f"Standard Deviation: {std_dev_val}")
```
