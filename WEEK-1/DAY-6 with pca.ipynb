```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset
```

```python
cancer_dataset.keys()
```

```python
type(cancer_dataset)
```

```python
cancer_dataset.DESCR
```

```python
df = pd.DataFrame(cancer_dataset.data, columns=cancer_dataset.feature_names)
```

```python
df["target"] = cancer_dataset.target
```

```python
df
```

```python
#Standardization

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data # [-3 to +3] and [mean=0 and sd=1]
```

```python
## Apply PCA Algorithm
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) # n_components means no. of components to be extracted
```

```python
data_pca=pca.fit_transform(scaled_data)
data_pca
```

```python
pca.explained_variance_ # variance of feature1 and feature2
```

```python
# to check with the help of these two features will we able to do something
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset["target"],cmap="plasma")
plt.xlabel("First principal component: ")
plt.ylabel("Second Principal Component: ")
```

```python
df.corr() # high correlation hints at collinearity
```

# Tsne Visualizaton

perplexity is used to determine whether o focus on local or glocal structures

```python
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
```

```python
data, labels = load_digits(return_X_y=True)
```

```python
data
```

```python
labels.shape
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.shape
```

```python
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
transformed_data = tsne.fit_transform(df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
```
